# Voos Nacionais e Internacionais de 2005 a 2017


Análise de dados voos rastreados pela Agência Nacional de Aviação Civil, no Brasil, de janeiro de 2015 a agosto de 2017.

Rankings com as cias e aeroportos mais utilizados.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
import os
```

```python
df = pd.read_csv('BrFlights2.csv', encoding='latin1')
```

```python
df.columns = ['Voos', 'Companhia_Aerea', 'Codigo_Tipo_Linha','Partida_Prevista','Partida_Real','Chegada_Prevista','Chegada_Real','Situacao_Voo','Codigo_Justificativa','Aeroporto_Origem','Cidade_Origem','Estado_Destino','Pais_Origem','Aeroporto_Destino','Cidade_Destino','Estado_Destino','Pais_Destino','LongDest','LatDest','LongOrig','LatOrig']
```

### Dados Analisados

```python
df.head()
```

### Tipo de rota mais utilizada

```python
df['Codigo_Tipo_Linha'].value_counts().plot.pie()
```

### Top 10 Cias Aéreas com voos com destino ao Estados Unidos

```python
plt.figure(figsize=(15,6))
plot = df[df.Pais_Destino == 'Estados Unidos'].Companhia_Aerea.value_counts().head(10).plot(kind="bar")
plot.set_title("Voos com destino ao Estados Unidos")
```

### Top Cia Aérea Nacional

```python
plt.figure(figsize=(15,6))
plot = df[df.Codigo_Tipo_Linha == 'Nacional'].Companhia_Aerea.value_counts().head(5).plot(kind="bar", )
plot.set_title("Top CIA Aerea Nacional")
```

### Top Cia Aérea Internacional

```python
plt.figure(figsize=(15,6))

plot = df[df.Codigo_Tipo_Linha == 'Internacional'].Companhia_Aerea.value_counts().head(5).plot(kind="bar")
plot.set_title("Top CIA Aerea Internacional")
```

### Voos Realizados e Cancelados

```python
df['Situacao_Voo'].value_counts().plot.pie()
```

### Cias que mais tiveram voos cancelados

```python
plt.figure(figsize=(15,6))
plot = df[df.Situacao_Voo == 'Cancelado'].Companhia_Aerea.value_counts().head(5).plot(kind="bar")
plot.set_title("CIA aerea com mais voos cancelados")
```

### Aeroportos mais utilizados como origem

```python
plt.figure(figsize=(15,6))
plot = df.Aeroporto_Origem.value_counts().head(10).plot(kind="bar")
plot.set_title("Top aeroportos origem")
```

### Cidades mais utilizadas como Origem

```python
plt.figure(figsize=(15,6))
plot = df.Cidade_Origem.value_counts().head(10).plot(kind="bar")
plot.set_title("Top cidades origem")
```

### Aeroportos mais utilizados como destino

```python
plt.figure(figsize=(15,6))
plot = df.Aeroporto_Destino.value_counts().head(10).plot(kind="bar")
plot.set_title("Top aeroportos destino")
```

### Cidades mais utilizadas como Destino

```python
plt.figure(figsize=(15,6))
plot = df.Cidade_Destino.value_counts().head(10).plot(kind="bar")
plot.set_title("Top cidades destino")
```
